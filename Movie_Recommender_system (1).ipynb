```python
import ast

def convert(obj):
    if isinstance(obj, str):  # If it's a string, parse it into a list
        obj = ast.literal_eval(obj)
    elif not isinstance(obj, list):  # If it's neither string nor list, return an empty list
        return []

    return [i['name'] for i in obj if isinstance(i, dict) and 'name' in i]  # Extract 'name' safely

```

```python
import ast

def convert3(obj):
    L = []

    # Handle missing or empty values
    if not obj or obj == "[]" or obj == "":
        return []

    # Convert string representation of a list to an actual list
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)  # Convert JSON-like string to list
        except (ValueError, SyntaxError):
            return []  # Return empty list if parsing fails

    # Ensure obj is a valid list before processing
    if not isinstance(obj, list):
        return []

    # Extract up to 3 'name' values safely
    for index, item in enumerate(obj):
        if index < 3 and isinstance(item, dict) and 'name' in item:
            L.append(item['name'])
        else:
            break  # Stop if we reach 3 items

    return L
```

```python
import ast

def fetch_director(obj):
    # Handle missing or empty values
    if not obj or obj == "[]" or obj == "":
        return []

    # Convert string representation of a list to an actual list
    if isinstance(obj, str):
        try:
            obj = ast.literal_eval(obj)  # Convert JSON-like string to list
        except (ValueError, SyntaxError):
            return []  # Return empty list if parsing fails

    # Ensure obj is a valid list before processing
    if not isinstance(obj, list):
        return []

    # Extract only the first director's name
    for item in obj:
        if isinstance(item, dict) and item.get('job') == 'Director' and 'name' in item:
            return [item['name']]  # Return as a list with a single name

    return []  # Return empty list if no director is found
```

```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('/content/tmdb_5000_movies.csv')
df2 = pd.read_csv('/content/tmdb_5000_credits.csv')

# Check their sizes
print(f"Movies Dataset: {df1.shape}")  # Should be (5000, X)
print(f"Credits Dataset: {df2.shape}")  # Should be (5000, X)
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# Rename 'movie_id' to 'id' in df2
df2.rename(columns={'movie_id': 'id'}, inplace=True)

# Now merge the datasets correctly
df = df1.merge(df2, on='id')
new_df = df[['id', 'title_x', 'overview']].copy()
new_df.rename(columns={'title_x': 'title', 'overview': 'tags'}, inplace=True)
```

```python
print(new_df.head())  # See first few rows
print(new_df.columns)  # Check available columns
```

```python
def recommend(movie):
    if movie not in new_df['title'].values:
        print("Movie not found! Please try another movie.")
        return

    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]

    # Sort based on similarity and remove zero similarity movies
    movies_list = sorted(list(enumerate(distance)), key=lambda x: x[1], reverse=True)[1:]
    movies_list = [m for m in movies_list if m[1] > 0]  # Remove zero similarities

    # Display top 5 results
    if not movies_list:
        print("No similar movies found.")
        return

    print(f"Top 5 recommendations for '{movie}':")
    for i in movies_list[:5]:
        print(new_df.iloc[i[0]].title)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Ensure 'tags' column exists in new_df and is a string
new_df['tags'] = new_df['tags'].astype(str)

# Convert text to vector using TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
vectorized_tags = tfidf.fit_transform(new_df['tags'])

# Compute cosine similarity
similarity = cosine_similarity(vectorized_tags)
```
